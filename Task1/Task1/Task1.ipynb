```python
import pandas as pd

insurance_df =  pd.read_csv('insurance.csv')
```

```python
insurance_df
```

```python
insurance_df['sex'] = insurance_df['sex'] == 'male'
insurance_df['smoker'] = insurance_df['smoker'] == 'yes'

insurance_df
```

```python
insurance_df = pd.get_dummies(insurance_df, columns=['region'], prefix=['region'])
insurance_df = insurance_df.astype(int)
insurance_df
```

```python
X = insurance_df.iloc[:,:-1]
y = insurance_df['charges']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
model = reg.fit(X_train,y_train)
train_accuracy = model.score(X_train,y_train)
test_accuracy = model.score(X_test,y_test)
print("train_accuracy: " + str(train_accuracy))
print("test_accuracy: "+ str(test_accuracy))

```

```python
def mse(y_true, y_pred):
    ########### write code here #############
    error = 0
    for i,y in enumerate(y_true):
        error += ((y-y_pred[i])**2)
    return error/len(y_pred)
```

```python
import numpy as np

# Linear Regression

class My_LinearRegression() :

    def __init__( self, learning_rate, iterations, type="linear" , reg_parameter = 0) :
        """
        This function initializes the object instance with arguments learning_rate, number of iterations, and whether to use linear or ridge or lasso
        reg_paramter is lambda for regularized regression"""

        self.learning_rate = learning_rate

        self.iterations = iterations

        self.type = type

        self.lam = reg_parameter

    def fit( self, X, Y ) :

        self.n = X.shape[0]
        self.m = X.shape[1]
        # weight initialization, initialize w to random numbers from a uniform distribution over [0,1[
        self.W = np.zeros(self.m)
        self.b = 0
        self.X = X
        self.Y = Y

        # gradient descent learning

        for i in range( self.iterations ) :
            self.update_weights()
        return self

    def update_weights( self, ) :

        Y_pred = (self.predict( self.X ))

        # calculate gradients
        dW = -2 * (1/self.n) * np.dot(self.X.T, self.Y - Y_pred )

        db = -2 * (1/self.n) * np.sum(self.Y - Y_pred)
        # dW = - ( 2 * ( self.X.T ).dot( self.Y - Y_pred )  ) / self.n

        # db = - 2 * np.sum( self.Y - Y_pred ) / self.n
        # update weights
        # print(self.W)
        print(dW)
        self.W -= self.learning_rate * dW

        self.b -= self.learning_rate * db

        # you don't need to return w and b
        return self

    def predict( self, X ) :
        y_pred = np.dot(X,self.W) + self.b

        return y_pred
```

```python
my_reg = My_LinearRegression(0.1,1000)
# print(X_train.to_numpy())
my_reg.fit(X_train.to_numpy(),y_train.to_numpy())
train_score = mse(my_reg.predict(X_train),y_train)
test_score = mse(my_reg.predict(X_test),y_test)
print('Train_Score: '+str(train_score))
print('Test_Score: '+str(test_score))
```
